### What is Panda?

- Pandas is a fast,powerful,flexible and easy to use open source data analysis and manipulation
tools,built on top of the python programming language.

### Panda Series

- A Pandas Series is like a column in a table. it is a 1-D array holding data of any type.

### Importing Panda

```python
import numpy as np
import pandas as pd
```

### Series from lists

```python
# String
country = ["India", "pakistan", "USA", "Nepal", "Srilanka"]
pd.Series(country)
```

```python
# integers
Runs = [32, 34, 56, 76, 100]
pd.Series(Runs)
```

```python
# Custom index
marks = [65, 85, 92, 88, 100]
subject = ["M-1", "BCE", "Physics", "BCEM", "BME"]
pd.Series(marks, index=subject)
```

```python
# setting a name
marks = [65, 85, 92, 88, 100]
subject = ["M-1", "BCE", "Physics", "BCEM", "BME"]
pd.Series(marks, index=subject, name="Ankit marksheet")
```

### Series from dict

```python
marks = {"maths": 67, "english": 57, "science": 89, "hindi": 100}
marks_series = pd.Series(marks, name="Ankit Markshet")
marks_series
```
